```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset
data = pd.read_csv('fetal_health.csv')
```

```python
# Separate features and target
X = data.drop(columns=['fetal_health'])
y = data['fetal_health']
```

```python
# Split the dataset into training and testing sets
# Using the first 75% rows for training and the last 25% for testing
split_index = int(0.75 * len(data))
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]
```

```python
# Initialize a baseline classifier ( We are using Random Forest)
classifier = RandomForestClassifier(random_state=42)
```

```python
# Train the classifier on the training data
classifier.fit(X_train, y_train)
```

```python
# Predict on the test data
y_pred = classifier.predict(X_test)
```

```python
# Compute and print the accuracy of the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"Baseline Model Accuracy: {accuracy:.2f}")
```
